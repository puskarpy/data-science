```python
norm_data = pd.DataFrame(np.random.normal(size = 100000))

norm_data.plot(kind="density",
               figsize=(10,10));

plt.vlines(norm_data.mean(),
           ymin=0,
           ymax=0.4,
           linewidth = 5.0,
           color = "green",
           );

plt.vlines(norm_data.median(),
           ymin=0,
           ymax=0.4,
           linewidth=2.0,
           color="red");
```
